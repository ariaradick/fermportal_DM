```julia
function make_histograms_xsec(run_info::RunsInfo, lumi, mass_idxs; 
                              binsize=[0.2, 0.5,
                              ustrip(natural(6.0*u"cm*ns^-1"))],
                              fig_idx=1:3,
                              orient="vertical",
                              text=nothing)
    N_mphis = length(mass_idxs)

    labels = Vector{String}()
    for m in (1e-3 .* run_info.mphis[mass_idxs])
        push!(labels, L"$m_\phi = %$(round(m,sigdigits=4))$ TeV")
    end

    u_summaries = get_run_summaries(run_info, mass_idxs; sfname=U_FILENAME)
    d_summaries = get_run_summaries(run_info, mass_idxs; sfname=D_FILENAME)
    
    N_events_u = N_events.(u_summaries)
    N_events_d = N_events.(d_summaries)

    stitle = L"$\phi_Q^{2/3}$ \& $\phi_Q^{1/3}$ Final States ($Q$ Model)"
    xlabels = [L"\beta \gamma", L"\eta", L"$L/\tau_{\phi}$ [cm/ns]", L"\beta"]
    colors = ["tab:blue", "tab:orange", "tab:red", "tab:cyan"]
    colorms = ["tab:blue", "tab:orange", "tab:red"]
    linestyles = ["solid", "dashed"]
    sizes_v = [(6,5), (6,8), (6,11)]
    sizes_h = [(6,5), (10,5), (15,5)]

    if orient == "vertical"
        fig, axs = subplots(length(fig_idx), 1, figsize=sizes_v[length(fig_idx)])
        sheight = .98
    elseif orient == "horizontal"
        fig, axs = subplots(1, length(fig_idx), figsize=sizes_h[length(fig_idx)])
        sheight = .85
    end

    u_binmax_ = maximum(maximum.(.(u_summaries)))
    d_binmax_ = maximum(maximum.(.(d_summaries)))
    binmax_ = max(u_binmax_, d_binmax_)
    _bins = range(0, ceil(binmax_/binsize[1])*binsize[1]; step=binsize[1])

    _bins = range(-7.5, 7.5; step=binsize[2])
    
    u_binmax_L = maximum(maximum.(L.(u_summaries)))
    d_binmax_L = maximum(maximum.(L.(d_summaries)))
    binmax_L = max(u_binmax_L, d_binmax_L)
    L_bins = range(0, ceil(binmax_L/binsize[3])*binsize[3]; step=binsize[3])

    beta_bins = 0.0:0.05:1.0

    binns = [_bins, _bins, L_bins, beta_bins]
    binns_scaled = [_bins, _bins, ustrip.(unnatural.(u"cm*ns^-1", L_bins)),
                    beta_bins]

    xmins = [0, -7.5, 0, 0]
    xmaxs = [_bins[end], 7.5, ustrip(unnatural(u"cm*ns^-1", L_bins[end])), 1]

    fns = [, , L, ]

    if length(fig_idx) == 1
        axss = (axs,)
    else
        axss = axs
    end

    for (i, ax) in enumerate(axss)
        minN = zeros(Float64, N_mphis)
        maxN = zeros(Float64, N_mphis)

        u_binned_stuff = bin_data.(fns[fig_idx[i]].(u_summaries), (binns[fig_idx[i]],))
        d_binned_stuff = bin_data.(fns[fig_idx[i]].(d_summaries), (binns[fig_idx[i]],))

        u_scaled_stuff = xsec.(u_summaries) .* lumi .* u_binned_stuff ./ N_events_u
        d_scaled_stuff = xsec.(d_summaries) .* lumi .* d_binned_stuff ./ N_events_d

        N_u_expected = sum(u_scaled_stuff; dims=1)
        N_d_expected = sum(d_scaled_stuff; dims=1)
        N_expected = N_u_expected .+ N_d_expected
        # println(sum.(N_expected))

        for j in eachindex(N_expected)
            minN[j] = minimum(N_expected[j][N_expected[j] .> 0])
            maxN[j] = maximum(N_expected[j])
            n = @. N_expected[j] + (N_expected[j] == 0)*1e-50 # trick to make stairs work
            ax.stairs(log10.(n), edges=binns_scaled[fig_idx[i]], 
                color=colorms[j], linestyle=linestyles[1])
        end

        # ylim_min = floor(log10(minimum(minN)))
        # ylim_max = ceil(log10(maximum(maxN)))

        ylim_min = 0
        ylim_max = log10(3e4)

        ax.set_xlabel(xlabels[fig_idx[i]])
        ax.set_ylabel("\\# of Expected Events")

        ax.grid(alpha=0.3)

        ax.set_xlim((xmins[fig_idx[i]], xmaxs[fig_idx[i]]))
        ax.set_ylim((ylim_min, ylim_max))

        x_extent = xmaxs[fig_idx[i]]-xmins[fig_idx[i]]
        y_extent = ylim_max - ylim_min

        if y_extent <= 5
            yticks = ylim_min:ylim_max
        else
            yticks = ylim_min:2:ylim_max
        end
        yticklabels = [L"10^{%$(round(Int,x))}" for x in yticks]
        ax.set_yticks(yticks, yticklabels)

        ax.set_aspect(x_extent / (y_extent*golden_ratio), adjustable="box")

    end

    if length(fig_idx) == 1
        for i in eachindex(labels)
            axs.plot(0, Inf, lw=1., c=colors[i], label=labels[i])
        end

        handles, labels = axs.get_legend_handles_labels()

        axs.set_title(stitle)

        axs.legend(handles, labels, loc="lower right", fontsize=12)

        props = Dict("boxstyle"=>"round", "facecolor"=>"white", "alpha"=>1)
        axs.text(0.025, 0.95, text, transform=axs.transAxes, 
                fontsize=12, verticalalignment="top", horizontalalignment="left",
                bbox=props)
    else
        for i in eachindex(labels)
            axs[0].plot(0, Inf, lw=1., c=colors[i], label=labels[i])
        end

        handles, labels = axs[0].get_legend_handles_labels()

        # axs[0].set_title(L"$\mu^+ \mu^- \to \phi^+ \phi^-$ ($\ell_R$ Model)")
        fig.suptitle(stitle, y=sheight)

        fig.legend(handles, labels, loc="lower center", ncol=4, fontsize=12)
    end

    tight_layout(rect=(0, 0.05, 1, 1))

    plotclose()
    return fig
end
;
```

## 2D Histogram $\beta \gamma$ vs $\eta$

```julia
function fbin(summary, _range, _range)
    Nevent = length(summary[1,:])
    _test = _range[1] .<= summary[2,:] .< _range[2]
    _test = _range[1] .<= summary[3,:] .< _range[2]
    res = sum(_test .&& _test)
    return res / Nevent
end

function bin_data_2d(rts_runs, lumi; binsize=[[0.05, 0.05, 0.05], 0.5])
    N_mphis = length(rts_runs.mphis)
    mphi_runs = get_run_with_summary.((rts_runs,), Vector(1:N_mphis))
    N_events = length(mphi_runs[1].mumu_summary[1,:])

    eachmax_ = [maximum(mphi_runs[j].mumu_summary[2,:]) for j in 1:N_mphis]
    eachmax2_ = [maximum(mphi_runs[j].vv_summary[2,:]) for j in 1:N_mphis]
    _locs = Vector{Vector{Float64}}()

    _bins = range(-5.5, 5.5; step=binsize[2])
    _locs = (_bins[2:end] .+ _bins[1:end-1])./2

    _s = [[_bins[i], _bins[i+1]] for i in 1:(length(_bins)-1)]

    res = Vector{Matrix{Float64}}()

    for i in 1:N_mphis
        binmax_ = max(eachmax_[i], eachmax2_[i])
        _bins = range(0, ceil(binmax_/binsize[1][i])*binsize[1][i]; 
                    step=binsize[1][i])
        push!(_locs, (_bins[2:end] .+ _bins[1:end-1])./2)

        _s = [[_bins[i], _bins[i+1]] for i in 1:(length(_bins)-1)]

        frac_mumu = fbin.((mphi_runs[i].mumu_summary,), _s, _s')
        frac_vv = fbin.((mphi_runs[i].vv_summary,), _s, _s')

        n_mumu = @. mphi_runs[i].mumu_xsec * lumi * frac_mumu
        n_vv = @. mphi_runs[i].vv_xsec * lumi * frac_vv

        push!(res, n_mumu .+ n_vv)
    end
    return (N_events=res, s=_locs, s=_locs, rts=rts_runs.rts,
            lumi=lumi, mphis=(1e-3 .* rts_runs.mphis))
end

function make_2d_hist(binned_data)
    title = L"$\sqrt{s} = %$(1e-3*binned_data.rts)$ TeV, $\mathcal{L} = %$(binned_data.lumi*1e-6)$ ab$^{-1}$"
    labels = Vector{String}()

    for m in binned_data.mphis
        push!(labels, L"$m_\phi = %$(round(m,digits=1))$ TeV")
    end

    xlabels = [L"\beta \gamma", L"\eta"]
    colors = ["tab:blue", "tab:orange", "tab:red"]
    linestyles = ["solid", "dashed"]

    fig, axs = subplots(3, 1, figsize=(5, 10))

    min_N = 1e-7
    max_N = 1e3

    L_levels = 10 .^ range(log10(.2), log10(to_cm_ns*maximum(maximum.(binned_data.s))), 8)
    formatter(l) = L"%$(round(l, sigdigits=2))"

    for (i, a) in enumerate(axs)
        a.set_xlim(-5.5, 5.5)
        a.set_ylim(0, binned_data.s[i][end] + 
            (binned_data.s[i][2]-binned_data.s[i][1])/2)
            
        g = a.pcolormesh(binned_data.s, binned_data.s[i], 
            binned_data.N_events[i], norm="log", cmap="inferno",
            vmin=min_N, vmax=max_N)

        _max = binned_data.s[i][end] + (binned_data.s[i][2]-binned_data.s[i][1])/2
        _L = range(0, _max, 50)
        s_L = range(-5.5, 5.5, 50)
        Ls = @. to_cm_ns * _L / cosh(s_L')
        C = a.contour(s_L, _L, Ls,
                    levels=L_levels, colors="black")

        txts = clabel(C, L_levels, colors="k", fmt=formatter, 
            use_clabeltext=true, inline=true)#, fontsize=16)
        
        for t in txts
            t.set_path_effects([pe.withStroke(linewidth=3, foreground="w")])
        end

        a.set_xlabel(L"\eta")
        a.set_ylabel(L"\beta \gamma")

        a.set_title(labels[i])

        a.grid(alpha=0.3)

        fig.colorbar(g)
    end

    fig.suptitle(title)

    plotclose()
    return fig
end
;
```

## 2D Histogram $L/\tau_\phi$ vs $\eta$

```julia
function fbin_2(summary, L_range, _range)
    Nevent = length(summary[1,:])
    L = summary[2,:] ./ cosh.(summary[3,:])
    L_test = L_range[1] .<= L .< L_range[2]
    _test = _range[1] .<= summary[3,:] .< _range[2]
    res = sum(L_test .&& _test)
    return res / Nevent
end

function bin_data_2d_L(rts_runs, lumi; binsize=[[0.05, 0.05, 0.05],.5])
    N_mphis = length(rts_runs.mphis)
    mphi_runs = get_run_with_summary.((rts_runs,), Vector(1:N_mphis))
    N_events = length(mphi_runs[1].mumu_summary[1,:])

    mumu_L = [mphi_runs[j].mumu_summary[2,:] ./ 
                cosh.(mphi_runs[j].mumu_summary[3,:]) for j in 1:N_mphis]
    vv_L = [mphi_runs[j].vv_summary[2,:] ./ 
            cosh.(mphi_runs[j].vv_summary[3,:]) for j in 1:N_mphis]

    eachmax_ = [maximum(mphi_runs[j].mumu_summary[2,:]) for j in 1:N_mphis]
    eachmax2_ = [maximum(mphi_runs[j].vv_summary[2,:]) for j in 1:N_mphis]
    L_locs = Vector{Vector{Float64}}()

    _bins = range(-5.5, 5.5; step=binsize[2])
    _locs = (_bins[2:end] .+ _bins[1:end-1])./2
    _s = [[_bins[i], _bins[i+1]] for i in 1:(length(_bins)-1)]

    res = Vector{Matrix{Float64}}()

    for i in 1:N_mphis
        binmax_ = max(eachmax_[i], eachmax2_[i])
        _bins = range(0, ceil(binmax_/binsize[1][i])*binsize[1][i]; 
                    step=binsize[1][i])
        push!(L_locs, (_bins[2:end] .+ _bins[1:end-1])./2)

        L_s = [[_bins[i], _bins[i+1]] for i in 1:(length(_bins)-1)]

        frac_mumu = fbin_2.((mphi_runs[i].mumu_summary,), L_s, _s')
        frac_vv = fbin_2.((mphi_runs[i].vv_summary,), L_s, _s')

        n_mumu = @. mphi_runs[i].mumu_xsec * lumi * frac_mumu
        n_vv = @. mphi_runs[i].vv_xsec * lumi * frac_vv

        push!(res, n_mumu .+ n_vv)
    end
    return (N_events=res, s=_locs, Ls=L_locs, rts=rts_runs.rts,
            lumi=lumi, mphis=(1e-3 .* rts_runs.mphis))
end

function make_2d_hist_L(binned_data)
    title = L"$\sqrt{s} = %$(1e-3*binned_data.rts)$ TeV, $\mathcal{L} = %$(binned_data.lumi*1e-6)$ ab$^{-1}$"
    labels = Vector{String}()

    for m in binned_data.mphis
        push!(labels, L"$m_\phi = %$(round(m,digits=1))$ TeV")
    end

    fig, axs = subplots(3, 1, figsize=(5, 10))

    min_N = 1e-7
    max_N = 1e3

    for (i, a) in enumerate(axs)
        L_scaled = ustrip.(unnatural.(u"cm*ns^-1", binned_data.Ls[i]))

        g = a.pcolormesh(binned_data.s, L_scaled, 
            binned_data.N_events[i][:,:], norm="log", 
            vmin=min_N, vmax=max_N, cmap="inferno")

        a.set_xlabel(L"\eta")
        a.set_ylabel(L"$L/\tau_\phi$ [cm/ns]")

        a.set_xlim(-5.5, 5.5)
        a.set_ylim(0, ceil(maximum(L_scaled)))

        a.set_title(labels[i])

        a.grid(alpha=0.3)

        fig.colorbar(g)
    end

    fig.suptitle(title)

    plotclose()
    return fig
end
;
```

```julia
function Lz(run_summary::RunSummary)
    (run_summary) .* tanh.((run_summary))
end

function events_in_detector_1(summary::RunSummary, lumi, tauphi, R_cm, Z_cm)
    pref = summary.xsec * lumi / N_events(summary)

    L = L(summary).*tauphi.*to_cm_ns
    Lz = Lz(summary).*tauphi.*to_cm_ns
    
    return pref*count((R_cm[1] .< L .< R_cm[2]) .&& (Lz .< Z_cm))
end

function events_in_detector_1(summary, lumi, tauphi, R_cm, _bound)
    pref = summary.xsec * lumi / N_events(summary)

    L = L(summary).*tauphi.*to_cm_ns
    etas = eta(summary)

    N_passes = count((R_cm[1] .< L .< R_cm[2]) .&& (abs.(etas) .< _bound))
    
    return pref*N_passes
end

function events_in_detector(run_info::RunsInfo, lumi, s, R_cm, Z_cm)
    N_mphi = length(run_info.mphis)
    N_mchi = length(s[1,:])

    res = zeros(Float64, (N_mphi, N_mchi))

    for i in 1:N_mphi
        summary = get_run_summaries(run_info, i)
        for j in 1:N_mchi
            if mchis[j] >= run_info.mphis[i]
                res[i,j] = NaN
            else
                N_expecteds = events_in_detector_1.(summary, lumi, s[i,j], (R_cm,), Z_cm)
                res[i,j] = sum(N_expecteds)
            end
        end
    end
    return res
end

function events_in_detector_(run_info::RunsInfo, lumi, s, R_cm, _bound)
    N_mphi = length(run_info.mphis)
    N_mchi = length(s[1,:])

    res = zeros(Float64, (N_mphi, N_mchi))

    for i in 1:N_mphi
        summary = get_run_summaries(run_info, i)
        for j in 1:N_mchi
            if mchis[j] >= run_info.mphis[i]
                res[i,j] = NaN
            else
                N_expecteds = events_in_detector_1.(summary, lumi, s[i,j], (R_cm,), _bound)
                res[i,j] = sum(N_expecteds)
            end
        end
    end
    return res
end

function events_in_detector(run_info::RunsInfo, lumi, mchis)
    detector_Rs = [[3, 10.4], [12.7, 55.4], [81.9, 148.6], [150, 170.2], 
                    [174, 333], [348.3, 429.0], [446.1, 645.0], [645.0, Inf]]
    detector_Zs = [65, (69.2+48.2)/2, 124.9, 221.0, 221.0, 412.9, 417.9, Inf]

    det_labs = ["in Vertex Detector", "in Inner Tracker", "in Outer Tracker", "in ECAL",
                "in HCAL", "in Solenoid", "in Muon Detector", "Outside"]

    N_exp = zeros(Float64, (length(detector_Rs), length(run_info.mphis), 
                    length(mchis)))
    
    s = lifetimes(run_info.mphis, mchis).*1e9
    
    for i in 1:(length(detector_Rs)-1)
        N_exp[i,:,:] = events_in_detector(run_info, lumi, s, detector_Rs[i],
            detector_Zs[i])
    end

    N_exp[end,:,:] = events_in_detector_(run_info, lumi, s, 
                        detector_Rs[end], 0.61)

    return (N_exp=N_exp, rts=10e3, lumi=lumi, mphis=run_info.mphis,
            mchis=mchis, labels=det_labs)
end
;
```

```julia
function t_mip(summary::RunSummary, R_cm)
    return R_cm .* cosh.(eta(summary)) .*(1 ./ beta(summary) .- 1) ./ to_cm_ns
end

function make_t_hist(run_info::RunsInfo, lumi, mass_idxs; R_cm=446.1, log_binsize=.1)
    N_mphis = length(run_info.mphis[mass_idxs])

    labels = Vector{String}()
    for m in (1e-3 .* run_info.mphis[mass_idxs])
        push!(labels, L"$m_\phi = %$(round(m,sigdigits=4))$ TeV")
    end

    run_summaries = get_run_summaries(run_info, mass_idxs)
    
    N_events = length.(.(run_summaries))

    colors = ["tab:blue", "tab:orange", "tab:red"]
    linestyles = ["solid", "dashed"]

    fig = figure(figsize=(5,5))
    ax = gca()

    t_vals = t_mip.(run_summaries, R_cm)
    
    binmin_t = minimum(minimum.(t_vals))
    binmax_t = maximum(maximum.(t_vals))

    min_t = floor(log10(binmin_t)/log_binsize)*log_binsize
    # max_t = ceil(log10(binmax_t)/log_binsize)*log_binsize
    max_t = 3

    t_bins = 10 .^ range(min_t, max_t; step=log_binsize)

    minN = zeros(Float64, N_mphis)
    maxN = zeros(Float64, N_mphis)

    binned_stuff = bin_data.(t_vals, (t_bins,))
    scaled_stuff = xsec.(run_summaries) .* lumi .* binned_stuff ./ N_events

    N_expected = sum(scaled_stuff; dims=1)

    for j in eachindex(N_expected)
        minN[j] = minimum(N_expected[j][N_expected[j] .> 0])
        maxN[j] = maximum(N_expected[j])
        n = @. N_expected[j] + (N_expected[j] == 0)*1e-50 # trick to make stairs work
        ax.stairs(log10.(n), edges=log10.(t_bins), 
            color=colors[j], linestyle=linestyles[1], label=labels[j])
    end

    ylim_min = floor(log10(minimum(minN)))
    ylim_max = ceil(log10(maximum(maxN)))

    # a.set_yscale("log")

    ax.set_xlabel(L"$\Delta t$ [ns]")
    ax.set_ylabel("\\# of Expected Events")

    ax.grid(alpha=0.3)

    ax.set_xlim((min_t, max_t))
    ax.set_ylim((ylim_min, ylim_max))

    x_extent = log10(binmax_t) - log10(binmin_t)
    y_extent = ylim_max - ylim_min

    if y_extent <= 5
        yticks = ylim_min:ylim_max
    else
        yticks = ylim_min:2:ylim_max
    end
    yticklabels = [L"10^{%$(round(Int,x))}" for x in yticks]
    ax.set_yticks(yticks, yticklabels)

    xticks = ceil(log10(binmin_t)):floor(log10(binmax_t))
    xticklabels = [L"10^{%$(round(Int,x))}" for x in xticks]
    ax.set_xticks(xticks, xticklabels)

    ax.set_aspect(x_extent / (y_extent*golden_ratio), adjustable="box")

    ax.legend(fontsize=12)

    # title(L"$\sqrt{s} = 10$ TeV, $\mathcal{L} = %$(lumi*1e-6)$ ab$^{-1}$")
    title(L"$u_R$ Model")

    tight_layout(rect=(0, 0, 1, 1))

    plotclose()
    return fig
end
;
```
